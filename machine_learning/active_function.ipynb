```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

# sigmoid函数

```python
def sigmoid(z):
    return 1.0 / (1 + np.exp(-z))

def deriv_sigmoid(z):
    return sigmoid(z)*(1 - sigmoid(z))

x = np.linspace(-5, 5, 100)
y = sigmoid(x)
d = deriv_sigmoid(x)
plt.plot(x, y, label='sigmoid')
plt.plot(x, d, 'r--', label='deriv_sigmoid')
plt.legend()
plt.show()
```

# Tanh函数

```python
def tanh(z):
    return (np.exp(z) - np.exp(-z))/(np.exp(z) + np.exp(-z))
def tanh_(z):
    return 2*sigmoid(2*z)-1


plt.plot(x, tanh(x), label='tanh')
plt.legend()
plt.show()
```

# Relu函数

```python
def relu(z):
    return max(0, z)

y = [relu(i) for i in x]
plt.plot(x, y, label='Relu')
plt.legend()
plt.show()
```

# softmax函数

```python
def softmax(z):
    s = np.exp(z)/sum(np.exp(z))
    return np.round(s, 3)
def loss_soft(p):
    l = -np.log(p)
    return l
    
y = softmax(x)
plt.plot(x, y, label='softmax')
plt.legend()
plt.show()
```
