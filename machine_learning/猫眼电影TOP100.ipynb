```python
import json
import requests
from multiprocessing import Pool
from requests.exceptions import RequestException
import re
import time
```

```python
def get_one_page(url):
    response = requests.get(url)
    if response.status_code == 200:
        return response.text
    return None

def parse_one_page(html):
    pattern = re.compile('<dd>.*?board-index.*?>(\d+)</i>.*?data-src="(.*?)".*?name"><a'
                         +'.*?>(.*?)</a>.*?star">(.*?)</p>.*?releasetime">(.*?)</p>'
                         +'.*?integer">(.*?)</i>.*?fraction">(.*?)</i>.*?</dd>',re.S)
    items = re.findall(pattern,html)
    for item in items:
        yield {
            'index':item[0],
            'image':item[1],
            'title':item[2],
            'actor':item[3].strip()[3:],
            'time':item[4].strip()[5:],
            'score':item[5] + item[6]
        }

def write_to_file(content):
    with open('/sdcard/1/txt/result.txt','a',encoding='utf-8') as f:
        f.write(json.dumps(content,ensure_ascii=False) + '\n')
        f.close()
def get_one_detail(html):
    pattern = re.compile('<dd>.*?board-index.*?>(\d+)</i>.*?data-src="(.*?)".*?name"><a'
                         +'.*?>(.*?)</a>.*?star">(.*?)</p>.*?releasetime">(.*?)</p>'
                         +'.*?integer">(.*?)</i>.*?fraction">(.*?)</i>.*?</dd>',re.S)
    items = re.findall(pattern,html)
    #items = list(items)
    detail_list = []
    for item in items:
        detail = {}
        #detail['index'] = item[0],
        #detail['image'] = item[1],
        detail['title'] = item[2],
        detail['actor'] = item[3].strip()[3:],
        detail['time'] = item[4].strip()[5:],
        detail['score'] = float(item[5] + item[6])
        print('正在爬取{}'.format(detail['title']))
        detail_list.append(detail)
    return detail_list
def write_to_csv():
    total_detail = []
    for i in range(10):
        url = 'http://maoyan.com/board/4?offset=' + str(i*10)
        html = get_one_page(url)
        detail = get_one_detail(html)
        total_detail.extend(detail)
    import pandas as pd
    df = pd.DataFrame(total_detail)
    df.to_csv('/sdcard/1/maoyan100.csv',index=True)
    print('保存csv成功!')
def main(offset):
    url = 'http://maoyan.com/board/4?offset=' + str(offset)
    html = get_one_page(url)
    for item in parse_one_page(html):
        print(item)
        write_to_file(item)
```

```python
if __name__ == '__main__':
    a = time.time()
    '''
    for i in range(10):
        main(i*10)
    '''
    write_to_csv()
    b = time.time()
    print('花费: {}秒'.format(b-a))
```
