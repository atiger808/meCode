```python
import requests
import re
import time
import json
from multiprocessing import Pool
import pandas as pd
from bs4 import BeautifulSoup as bs
url1 = 'http://hz.58.com/ershoufang/pn1/'
```

```python
def parse(url):
    #headers = {'Server': 'Tengine', 'Date': 'Mon, 23 Apr 2018 04:44:49 GMT', 'Content-Type': 'text/html;charset=UTF-8', 'Transfer-Encoding': 'chunked', 'Connection': 'keep-alive', 'X-HOST': 'firewallverifycode', 'Content-Encoding': 'gzip', 'Vary': 'Accept-Encoding'}
    headers = {
'user-agent':'Mozilla/5.0 (Windows NT 10.0; WOW64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/60.0.3112.101 Safari/537.36',
'upgrade-insecure-requests':'1',
}
    res = requests.get(url, headers=headers)
    if res.status_code == 200:
        res.encoding = 'utf-8'
        return res.text
    return None
```

```python
def get_one_page_list(n, url):
    res = parse(url)
    if res:
        soup = bs(res, 'html.parser')
        li_list = soup.select('.house-list-wrap li')
        house_list = []
        for i in li_list:
            house_detail = {}
            try:
                house_detail['name'] = i.select('a')[2].text
                house_detail['types'] = ' '.join(i.select('span')[1].text.split())                      
                house_detail['area'] = i.select('span')[2].text.strip()
                house_detail['price'] = i.select('.unit')[0].text                                        
                house_detail['total-price'] = i.select('.sum')[0].text.strip()                                       
                house_detail['floor'] = i.select('span')[3].text
                house_detail['location'] = i.select('a')[4].text
                house_detail['link'] = i.select('a')[1]['href']
                n = n+1
                house_list.append(house_detail)
            except:
                continue
    return house_list, n
```

```python
def main(n):
    house_all_list = []
    for i in range(1, 71):
        url = 'http://hz.58.com/ershoufang/pn'+str(i)+'/'
        house_list, n = get_one_page_list(n, url)
        house_all_list.extend(house_list)
        print('正在爬取第%d页'%(i))
    print('完毕')
    return house_all_list
```

```python
total_data = main(1)
```

```python
df = pd.DataFrame(total_data)
```
